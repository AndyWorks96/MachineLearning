# 创建随机测试集

```python
 import numpy as np
```

```python
def split_train_test(data,test_ratio):
    sorted = np.random.permutation(len(data))
    test_size = (int)(len(data)*test_ratio)
    test_set = sorted[:test_size]
    train_set = sorted[test_size:]
    return test_set,train_set
```

```python
test_set,train_set = split_train_test(housing,0.2)
# print(train_set,"train",test_set,"test")
print(len(train_set),"train","+",len(test_set),"test")
```

# 加入哈希的测试集

```python
 np.random.seed(42)# 生成相同的随机索引
test_indices= np.random.permutation(5)
print(test_indices)
```

```python
import hashlib # hashlib是涉及安全散列和消息摘要，提供多个不同的加密算法接口
def test_set_check(identifier,test_ratio,hash=hashlib.md5):
    if hash(np.int64(identifier)).digest()[-1] < 256*test_ratio:
        print(hash(np.int64(identifier)).digest()[-1])
    return hash(np.int64(identifier)).digest()[-1] < 256*test_ratio
def split_test_set_by_id(data,id_columns,test_ratio):
    ids = data[id_columns]
    test_setId= ids.apply(lambda id_:test_set_check(id_,test_ratio))
    return data.loc[test_setId],data.loc[~test_setId]
```

```python
housing_with_id=housing.reset_index()
# print(housing_with_id)此处的housing_with_id
test_set,train_set=split_test_set_by_id(housing_with_id,"index",0.2)
```

```python
housing_with_id["id"] = housing["longitude"]*1000+housing["latitude"]# 增加一列命名为id
# print(housing_with_id)
test_set,train_set = split_test_set_by_id(housing_with_id,"id",0.2)
```

# sklearn提供的创建测试集的包

```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(housing,test_size=0.2,random_state=42)
print(len(train_set))
```

```python
housing["icome_cat"]=np.ceil(housing["median_income"]/1.5)# 划分类别
# print(housing)
housing["icome_cat"].where(housing["icome_cat"]<5,5.0,inplace=True)
print(housing)
```

# 根据收入类别进行分层抽样

```python
# 使用对象的split方法分割，返回的是分组后数在原数组中的索引
```

```python
from sklearn.model_selection import StratifiedShuffleSplit #提供分层抽样功能
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
# random_state：随机数种子，和random中的seed种子一样，保证每次抽样到的数据一样，便于调试
num = 0
for train_index,test_index  in split.split(housing,housing["icome_cat"]):
    # train_set = housing[train_index]
    # test_set = housing[test_index]错误
    
    num = num+1
    start_train_set = housing.loc[train_index]
    start_test_set = housing.loc[test_index]
print(start_train_set)
print(start_test_set)
```

```python
# 此处就是收入类别比例分布
housing["icome_cat"].value_counts()/len(housing)
```

```python
# 删除icome_cat属性，将数据恢复原样
for set in (start_train_set,start_test_set):
    set.drop(["icome_cat"],axis=1,inplace=True)
print(start_train_set)
```
